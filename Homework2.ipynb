### Tarea 1: Mapas estáticos: recuento de hospitales por distrito
* 
Cargue el shapefile del distrito de Perú (enlace arriba).* 
Con GeoPandas , crea mapas estáticos que muestren el número de hospitales públicos operativos por distrito- :
Mapa 1: Total de hospitales públicos por distrit- o.
Mapa 2: Resalte los distritos con cero hospital- es.
Mapa 3: Los 10 distritos con mayor número de hospitales (escala de colores diferenciada).

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# =========================
# 1. Leer hospitales y filtrar operativos
# =========================
hosp = pd.read_csv("IPRESS.csv", encoding="latin1")
hosp_op = hosp[(hosp["Condición"] == "EN FUNCIONAMIENTO") & 
               (~hosp["NORTE"].isna()) & 
               (~hosp["ESTE"].isna())]

# =========================
# 2. Crear GeoDataFrame de hospitales (EPSG:4326 recomendado)
# =========================
gdf_hosp = gpd.GeoDataFrame(
    hosp_op,
    geometry=gpd.points_from_xy(hosp_op["NORTE"], hosp_op["ESTE"]),
    crs="EPSG:4326"
)

# =========================
# 3. Leer distritos (EPSG:4326)
# =========================
dist = gpd.read_file("DISTRITOS.shp").to_crs("EPSG:4326")

# =========================
# 4. Spatial join: asignar hospitales a distritos
# =========================
hosp_distr = gpd.sjoin(gdf_hosp, dist, how="left", predicate="within")

# =========================
# 5. Conteo hospitales por distrito
# =========================
hosp_count = hosp_distr.groupby("IDDIST").size().reset_index(name="n_hosp")

# =========================
# 6. Unir conteo a distritos
# =========================
dist_hosp = dist.merge(hosp_count, on="IDDIST", how="left")
dist_hosp["n_hosp"] = dist_hosp["n_hosp"].fillna(0)

# =========================
# MAPA 1: Total hospitales por distrito
# =========================

fig, ax = plt.subplots(1, 1, figsize=(10, 10))
dist_hosp.plot(column="n_hosp", cmap="YlGnBu", legend=True, ax=ax, edgecolor="black")
plt.title("Mapa 1: Total de hospitales públicos operativos por distrito", fontsize=14)
plt.axis("off")
plt.tight_layout()

# Guardar primero
plt.savefig("output/Coropleta_Distrital.png", dpi=150)

# Mostrar en notebook (opcional, solo si quieres verlo ahí)
plt.show()

plt.close()
```

```python
# MAPA 2: Distritos con cero hospitales
# =========================
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
dist_hosp.plot(ax=ax, color="lightgrey")
dist_hosp[dist_hosp["n_hosp"] == 0].plot(ax=ax, color="red", edgecolor="black")
plt.title("Mapa 2: Distritos con cero hospitales", fontsize=14)
plt.axis("off")

plt.savefig("output/distritos_sin_hosp.png", dpi=300, bbox_inches="tight")
plt.show()

```

```python
# MAPA 3: Top 10 distritos con más hospitales
# =========================
top10 = dist_hosp.nlargest(10, "n_hosp")
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
dist_hosp.plot(ax=ax, color="lightgrey")
top10.plot(column="n_hosp", cmap="OrRd", legend=True, ax=ax, edgecolor="black")
plt.title("Mapa 3: Top 10 distritos con mayor número de hospitales", fontsize=14)
plt.axis("off")

plt.savefig("output/top10_distritos.png", dpi=300, bbox_inches="tight")
plt.show()
```

#### Tarea 3: Análisis de proximidad (utilizando centros de población)

```python
import pandas as pd
import geopandas as gpd
import folium


ccpp = gpd.read_file("CCPP_IGN100K.shp").to_crs(epsg=4326)
# =========================
# 1. Asegurar CRS correcto
# =========================
# Trabajamos en CRS métrico para calcular radios de 10 km
gdf_hosp_utm = gdf_hosp.to_crs(epsg=32718)   # hospitales operativos
ccpp_utm = ccpp.to_crs(epsg=32718)           # centros poblados (centroides)

# =========================
# 2. Filtrar por regiones (Lima y Loreto)
# =========================
ccpp_lima = ccpp_utm[ccpp_utm["DEP"] == "LIMA"].copy()
ccpp_loreto = ccpp_utm[ccpp_utm["DEP"] == "LORETO"].copy()

# =========================
# 3. Crear buffers de 10 km
# =========================
ccpp_lima["buffer"] = ccpp_lima.geometry.buffer(10000)      # 10 km = 10,000 m
ccpp_loreto["buffer"] = ccpp_loreto.geometry.buffer(10000)

# =========================
# 4. Contar hospitales dentro del buffer
# =========================
def contar_hosp(ccpp_gdf, hosp_gdf):
    conteos = []
    for geom in ccpp_gdf["buffer"]:
        conteos.append(hosp_gdf.within(geom).sum())
    return conteos

ccpp_lima["n_hosp"] = contar_hosp(ccpp_lima, gdf_hosp_utm)
ccpp_loreto["n_hosp"] = contar_hosp(ccpp_loreto, gdf_hosp_utm)

# =========================
# 5. Identificar aislados y concentrados
# =========================
# Lima
lima_min = ccpp_lima.loc[ccpp_lima["n_hosp"].idxmin()]
lima_max = ccpp_lima.loc[ccpp_lima["n_hosp"].idxmax()]

# Loreto
loreto_min = ccpp_loreto.loc[ccpp_loreto["n_hosp"].idxmin()]
loreto_max = ccpp_loreto.loc[ccpp_loreto["n_hosp"].idxmax()]

print("Lima - menos hospitales:", lima_min["NOM_POBLAD"], "-", lima_min["n_hosp"])
print("Lima - más hospitales:", lima_max["NOM_POBLAD"], "-", lima_max["n_hosp"])
print("Loreto - menos hospitales:", loreto_min["NOM_POBLAD"], "-", loreto_min["n_hosp"])
print("Loreto - más hospitales:", loreto_max["NOM_POBLAD"], "-", loreto_max["n_hosp"])

# =========================
# 6. Visualizar con Folium
# =========================
def mapa_buffer(cpp, hosp_gdf, nombre):
    # Pasamos todo a EPSG:4326 para que funcione en Folium
    cpp4326 = gpd.GeoDataFrame([cpp], geometry=[cpp.geometry], crs=32718).to_crs(4326)
    buffer4326 = gpd.GeoDataFrame(geometry=[cpp["buffer"]], crs=32718).to_crs(4326)
    hosp4326 = hosp_gdf.to_crs(4326)

    # Crear mapa centrado en el centro poblado
    m = folium.Map(location=[cpp4326.geometry.y.iloc[0], cpp4326.geometry.x.iloc[0]],
                   zoom_start=10, tiles="cartodbpositron")

    # Añadir buffer
    folium.GeoJson(buffer4326.geometry.iloc[0],
                   style_function=lambda x: {"color": "blue", "weight": 2, "fill": False}).add_to(m)

    # Centro poblado
    folium.Marker(
        [cpp4326.geometry.y.iloc[0], cpp4326.geometry.x.iloc[0]],
        popup=f"{cpp['NOM_POBLAD']}<br>Hospitales: {cpp['n_hosp']}",
        icon=folium.Icon(color="red" if cpp['n_hosp']==0 else "green")
    ).add_to(m)

    # Hospitales dentro del buffer
    hosp_dentro = hosp4326[hosp4326.within(buffer4326.geometry.iloc[0])]
    for _, row in hosp_dentro.iterrows():
        folium.CircleMarker(
            [row.geometry.y, row.geometry.x],
            radius=3, color="blue", fill=True, fill_opacity=0.7
        ).add_to(m)

    m.save(nombre)
    return m

# Generar mapas para Lima
mapa_buffer(lima_min, gdf_hosp_utm, "Lima_min.html")
mapa_buffer(lima_max, gdf_hosp_utm, "Lima_max.html")

# Generar mapas para Loreto
mapa_buffer(loreto_min, gdf_hosp_utm, "Loreto_min.html")
mapa_buffer(loreto_max, gdf_hosp_utm, "Loreto_max.html")

print("Mapas generados: Lima_min.html, Lima_max.html, Loreto_min.html, Loreto_max.html")
```

# 3) Mapeo interactivo con Folium

```python
import folium
from folium.plugins import MarkerCluster

# Asegurar CRS correcto
dist_hosp = dist_hosp.to_crs(4326)
gdf_hosp = gdf_hosp.to_crs(4326)

# Calcular centroide nacional a partir de la unión de distritos
centro = dist_hosp.geometry.union_all().centroid
centro_coords = [centro.y, centro.x]

# Crear mapa centrado en el centroide del país (derivado de tus datos)
m = folium.Map(location=centro_coords, zoom_start=5, tiles="cartodbpositron")

# 1) Coropleta por número de hospitales en distritos
folium.Choropleth(
    geo_data=dist_hosp.to_json(),
    name="Hospitales por distrito",
    data=dist_hosp,
    columns=["IDDIST", "n_hosp"],
    key_on="feature.properties.IDDIST",
    fill_color="YlGnBu",
    fill_opacity=0.7,
    line_opacity=0.3,
    legend_name="Número de hospitales en funcionamiento"
).add_to(m)

# 2) Marcadores de hospitales
marker_cluster = MarkerCluster(name="Hospitales (puntos)").add_to(m)

for _, row in gdf_hosp.iterrows():
    folium.Marker(
        location=[row.geometry.y, row.geometry.x],
        popup=f"{row['Nombre del establecimiento']}<br>{row['Distrito']}, {row['Provincia']}, {row['Departamento']}"
    ).add_to(marker_cluster)

# 3) Control de capas
folium.LayerControl(collapsed=False).add_to(m)

# Guardar
m.save("Coropleta_Distrital.html")
m
```

```python
# 1. Creamos función para mapas de proximidad

def mapa_buffer(cpp, hosp_gdf, nombre, color):
    # Pasamos a EPSG:4326 para visualización en Folium
    cpp4326 = gpd.GeoDataFrame([cpp], geometry=[cpp.geometry], crs=32718).to_crs(4326)
    buffer4326 = gpd.GeoDataFrame(geometry=[cpp["buffer"]], crs=32718).to_crs(4326)
    hosp4326 = hosp_gdf.to_crs(4326)

    # Crear mapa centrado en el centro poblado
    m = folium.Map(
        location=[cpp4326.geometry.y.iloc[0], cpp4326.geometry.x.iloc[0]],
        zoom_start=10, tiles="cartodbpositron"
    )

    # Añadir buffer (círculo de 10 km)
    folium.GeoJson(
        buffer4326.geometry.iloc[0],
        style_function=lambda x: {"color": color, "weight": 2, "fill": False}
    ).add_to(m)

    # Centro poblado con popup
    folium.Marker(
        [cpp4326.geometry.y.iloc[0], cpp4326.geometry.x.iloc[0]],
        popup=f"<b>{cpp['NOM_POBLAD']}</b><br>Hospitales dentro de 10 km: {cpp['n_hosp']}",
        icon=folium.Icon(color=color, icon="info-sign")
    ).add_to(m)

    # Hospitales dentro del buffer (azules)
    hosp_dentro = hosp4326[hosp4326.within(buffer4326.geometry.iloc[0])]
    for _, row in hosp_dentro.iterrows():
        folium.CircleMarker(
            [row.geometry.y, row.geometry.x],
            radius=3, color="blue", fill=True, fill_opacity=0.7
        ).add_to(m)

    # Guardar y devolver
    m.save(nombre)
    return m

# 2. Generamos mapas
# Lima
mapa_lima_min = mapa_buffer(lima_min, gdf_hosp_utm, "Lima_min.html", "red")     # menos hospitales
mapa_lima_max = mapa_buffer(lima_max, gdf_hosp_utm, "Lima_max.html", "green")   # más hospitales

# Loreto
mapa_loreto_min = mapa_buffer(loreto_min, gdf_hosp_utm, "Loreto_min.html", "red")
mapa_loreto_max = mapa_buffer(loreto_max, gdf_hosp_utm, "Loreto_max.html", "green")

print("Mapas generados: Lima_min.html, Lima_max.html, Loreto_min.html, Loreto_max.html")
```
